```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Admin\Downloads\_nfhw4ldp.0sc.csv')
# input your path and name of csv
```

```python
df.head()
```

# Get rid of lognom data

```python
df = df.drop(filter(lambda c : c.endswith('_lognorm'), df.columns), axis = 1)
```

```python
df.head()
```

```python
df['published'] = pd.to_datetime(df.published, yearfirst=True)
```

```python
df.head()
```

```python
sns.set_style('dark')
sns.set_palette('RdBu')
sns.set_context('notebook', font_scale=1.5,
               rc = {'figure.figsize': (15, 5), 'axes.titlesize': 18})
```

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]
df['day'] = [d.day for d in df.published]
df['day_of_week'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
df['month_year'] = df['published'].dt.to_period('M')
```

# What was the best month by publishing articles?  Lets find out

```python
df.groupby('month_year')[['title']].count().sort_values('title', ascending=False).head()
```

<b>So we see that in march 2015 were more articles
